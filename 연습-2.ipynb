```python
#문자열 연산
sen='  dgsw high school'

print(len(sen))
print(sen.count('s'))
print(sen.find('s')) #값이 존재 하지 않으면 -1 반환
print(sen.index('h')) #값이 존재 하지 않으면 오류
print(sen.upper())
print(sen.lower())
print(sen.lstrip())
print(sen.replace('dgsw','DGSW'))
print(sen.split(' '))
```
