```python
# 입력 받은 점수 70점 이상 합격, 그외 재도전, 90점이상은 장학금 대상

a=int(input('점수:'))

if a>=70:
    print('합격')
    if a>=90:
        print('장학금 대상') 
else:
    print('재도전')
```
