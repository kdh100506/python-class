```python
def sum(n):
  s=0
  for i in range(1, n+1):
    s+=i
  return s
print(sum(10))
```

```python
def pac(n):
  s=1
  for i in range(1,n+1):
    s*=i
  return s
print(pac(5))
```

```python
def com(a,b):
  if a>b:
    return a
  elif a<b:
    return b
  
a,b=map(int,input('수 입력').split())
print(f'큰수 : {com(a,b)}')
```

```python
width=int(input())
height=int(input())
def triangle(width,height):
  area=width*height/2
  return area
area=triangle(width,height)
print(f'삼각형의 넓이: {area}')
```

```python
hour=int(input())
def HtoS(hour):
  hour=hour*60*60
  return hour
second=HtoS(hour)
print(f'{hour}시간은 {second}초 입니다.')
```

```python
def list_avg(score):
  hap=0
  for i in score:
    hap+=i
  list_len=len(score)
  avg=hap/list_len
  print(avg)
score=[90,80,70,50,60]
list_avg(score)
```
