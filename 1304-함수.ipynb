```python
def sum(n):
  s=0
  for i in range(1, n+1):
    s+=i
  return s
print(sum(10))
```

```python
def pac(n):
  s=1
  for i in range(1,n+1):
    s*=i
  return s
print(pac(5))
```

```python
def com(a,b):
  if a>b:
    return a
  elif a<b:
    return b
  
a,b=map(int,input('수 입력').split())
print(f'큰수 : {com(a,b)}')
```
