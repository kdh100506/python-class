```python
#1번

words=input().lower().split()
gather='aeiou'
count=0

for i in words:
  for j in i:
    if j in gather:
      count+=1
print(f'모음 개수: {count}')
```

```python
#2번

nums=map(int,input().split())
even=0
odd=0

for i in nums:
  if i%2==0:
    even+=i
  else:
    odd+=i
print(f'짝수 합: {even}')
print(f'홀수 합: {odd}')
```

```python
#6번

password=input()
isalpha_count=0
isdigit_count=0

for i in password:
  if i.isalpha():
    isalpha_count+=1
  elif i.isdigit():
    isdigit_count+=1

if isalpha_count>=1 and isdigit_count >=1:
  if len(password)>=8:
    print('사용 가능')
  else:
    print('사용 불가')
else: print('사용 불가')
    

```
