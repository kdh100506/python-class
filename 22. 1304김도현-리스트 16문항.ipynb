```python
#1번

words=input().lower().split()
gather='aeiou'
count=0

for i in words:
  for j in i:
    if j in gather:
      count+=1
print(f'모음 개수: {count}')
```

```python
#4번

word=input()
word_list_1=[]
word_list_2=[]
word_len=len(word)

for i in range(word_len):
  word_list_1.append(word[i])

for i in range(word_len-1,-1,-1):
  word_list_2.append(word[i])

if word_list_1==word_list_2:
  print('회문 입니다')
else:
  print('회문이 아닙니다')
```

```python
#6번

password=input()
isalpha_count=0
isdigit_count=0

for i in password:
  if i.isalpha():
    isalpha_count+=1
  elif i.isdigit():
    isdigit_count+=1

if isalpha_count>=1 and isdigit_count >=1:
  if len(password)>=8:
    print('사용 가능')
  else:
    print('사용 불가')
else: print('사용 불가')
    

```

```python
#8번

student=int(input())
point_list=[]

for i in range(student):
  point=int(input())
  point_list.append(point)

point_len=len(point_list)
point_hap=0
count=0

for i in point_list:
  point_hap+=i
  if i>=80:
    count+=1

print(f'평균: {point_hap/point_len}')
print(f'합격자 수: {count}')
```

```python
#10번

people=int(input())
student={}

for i in range(people):
  name,point=input().split()
  student[name]=int(point)

student_len=len(student)
student_hap=0
for i in student:
  student_hap+=student[i]

for i in student:
  if student[i]>=student_hap/student_len:
    print(i, student[i])
```

```python
#11번

files=input().split()
py_count=0
txt_count=0
jpg_count=0

for i in files:
  if '.py' in i:
    py_count+=1
  elif '.txt' in i:
    txt_count+=1
  elif '.jpg' in i:
    jpg_count+=1
print('py',py_count)
print('txt', txt_count)
print('jpg', jpg_count)
```

```python
#12번

words=input()
words_one=set(words)
words_one=list(words_one)
words_one.sort()
words_dict={}

for i in words_one:
  words_dict[i]=words.count(i)

for i in words_dict:
  print(f'{i}{words_dict[i]}', end='')
```

```python
#13번

nums= list(map(int,input().split()))
nums_copy=nums.copy()
print(nums_copy)
count=int(input())
nums_len=len(nums)

for i in range(nums_len):
  if i+count >= nums_len:
    print(i-3, 'a')
    nums[i-3]=nums_copy[i]
  else:
    print(i+2, 'b')
    nums[i+2]=nums_copy[i]

print(nums)
```
