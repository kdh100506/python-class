```python
import random, time

w=['cat','dog','fox','monkey','mouse','panda','frog','snake','wolf']
ww=[]
n=0
i=0

print(f'타자게임: 준비되면 엔터')
input()
print()

start=time.time()
while i<5:
    i+=1
    print(f'문제 {i}')
    q=''
    q=random.choice(w)
    print(q)
    ans=input()
    if ans==q:
        print('pass')
        n+=1
    else:
        print('fail')
    print()

end=time.time()
re=end-start

print(f'맞은 개수: {n}')
print(f'걸린 시간: {re:.4f}초')
```

```python
import random, time

w=['cat','dog','fox','monkey','mouse','panda','frog','snake','wolf']
ww=[]
n=0
i=0

print(f'타자게임: 준비되면 엔터')
input()
print()

start=time.time()
while n<5:
    q=''
    q=random.choice(w)
    ww.append(q) #q를 ww리스트에 추가
    if ww.count(str(q))>1: #q가 ww리스트에 1보다 많으면 다시 처음으로
        continue
    i+=1
    print(f'문제 {i}')
    print(q)
    ans=input()
    if ans==q:
        print('pass')
        n+=1
    else:
        print('fail')
    print()

end=time.time()
re=end-start

print(f'맞은 개수: {n}')
print(f'걸린 시간: {re:.4f}초')
```
