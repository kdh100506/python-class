```python
import random as r

while True:
    start=r.randint(0,10)
    end=r.randint(start,10)
    hap=0
    if start==end:
        break

    for i in range(start,end+1):
        hap+=i
    print(f'{start}~{end} -> {hap}')
```

```python
hap=0
cnt=0

while True:
    score=input()
    if score=='q':
        break
    hap+=float(score)
    cnt+=1

print(f'합계: {hap}')
if hap>0:
    avg=hap/cnt
    print(f'평균: {avg}')
```

```python
sum=0
count=0
while True:
    n=input()
    if not n.replace('.','',1).isdigit():
        print('입력 종료')
        break
    count+=1
    sum+=float(n)
    
if count>0:
    print(f'합계: {sum}')
    print(f'평균: {sum/count}')
else:
    print('입력한 수가 없음')
```

```python
import random, time

w=['cat','dog','fox','monkey','mouse','panda','frog','snake','wolf']
n=0

start=time.time()

for i in range(5):
    q=''
    q=random.choice(w)
    print(f'문제 {i+1}')
    print(q)
    ans=input()
    if q==ans:
        print('pass')
        n+=1
    else:
        print('fail')
    print()
end=time.time()
re=end-start

print(f'맞은 개수: {n}')
print(f'걸린 시간: {re:.4f}초')
```

```python
import random, time

w=['cat','dog','fox','monkey','mouse','panda','frog','snake','wolf']
ww=[]
n=0
i=0

print(f'타자게임: 준비되면 엔터')
input()
print()

start=time.time()
while n<5:
    q=''
    q=random.choice(w)
    ww.append(q) #q를 ww리스트에 추가
    if ww.count(str(q))>1: #q가 ww리스트에 1보다 많으면 다시 처음으로
        continue
    i+=1
    print(f'문제 {i}')
    print(q)
    ans=input()
    if ans==q:
        print('pass')
        n+=1
    else:
        print('fail')
    print()

end=time.time()
re=end-start

print(f'맞은 개수: {n}')
print(f'걸린 시간: {re:.4f}초')
```
