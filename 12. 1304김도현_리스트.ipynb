```python
list1=[1,2,3,4,5]
list1.remove(3)
print(list1)

list1.pop() #()에 아무것도 안쓰면 마지막 리스트가 지워짐
print(list1)
list1.pop(1)
print(list1)

list1.reverse() #반전 시키기
print(list1)
```

```python
applicant=['홍길동','일지매']

applicant_append=input('추가할 학생 이름:')
applicant.append(applicant_append)
print(applicant_append,'학생의 수강 신청이 완료 되었습니다.')
print('현재 이 과목의 최종 신청자는',applicant,'입니다.')

applicant_pop=input('철회할 학생 이름:')
applicant.remove(applicant_pop)
print(applicant_pop,'학생의 수강 철회가 완료되었습니다.')
print('현재 이 과목의 최종 신청자는',applicant,'입니다.')

name_change_before=input('변경 전 이름:')
name_change_after=input('변경 후 이름:')
name_index=applicant.index(name_change_before)
applicant[name_index]=name_change_after
print('요청하신 대로',name_change_before+'을(를)',name_change_after+'(으)로 변경하였습니다.')
print('현재 이 과목의 최종  신청자는',applicant,'입니다.')
```
