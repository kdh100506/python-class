```python
persons=[
  {'name':'aaa','address':'Deagu','interset':'web'},
  {'name':'bbb','address':'Seoul','interset':'IOT'},
  {'name':'ccc','address':'LA','interset':'web'}
]

print('==== persons ====')
print('name:', persons[0]['name'])
print('address:', persons[0]['address'])
print('interset:', persons[0]['interset'])
print('-----------------')
print('name:', persons[1]['name'])
print('address:', persons[1]['address'])
print('interset:', persons[1]['interset'])
print('-----------------')
print('name:', persons[2]['name'])
print('address:', persons[2]['address'])
print('interset:', persons[2]['interset'])
print('-----------------')

print()

print('==== persons ====')
for person in persons:
  for key in person:
    print(f'{key}: {person[key]}') #key와 value를 출력하는 방법
  print('-----------------')
```

```python
#21

score={'민수':95,'지영':82,'철수':68}
grade={}

for i in score:
  if score[i]>=90:
    grade[i]='A'
  elif score[i]>=80:
    grade[i]='B'
  else:
    grade[i]='C'
print(score)
print(grade)
```
