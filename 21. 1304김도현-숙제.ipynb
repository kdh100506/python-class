```python
#12

password=input()
alpha='qwertyuiopasdfghjklzxcvbnmQWERTYUIOPASDFGHJKLZXCVBNM'
alpha_count=0
digit='1234567890'
digit_count=0
etc_count=0
password_dict={'alpha':0,'digit':0,'spacial':0}

for i in password:
  if i in alpha:
    alpha_count+=1
    password_dict['alpha']=alpha_count
  elif i in digit:
    digit_count+=1
    password_dict['digit']=digit_count
  else:
    etc_count+=1
    password_dict['spacial']=etc_count

print(password_dict)
```

```python
#18번

people=int(input())
student={}
A_count=0
B_count=0
C_count=0
D_count=0

for i in range(people):
  name,point=input().split()
  if int(point)>=90:
    if 'A' not in student:
      student['A']=0
    A_count+=1
    student['A']=A_count
  elif int(point)>=80:
    if 'B' not in student:
      student['B']=0
    B_count+=1
    student['B']=B_count
  elif int(point)>=70:
    if 'C' not in student:
      student['C']=0
    C_count+=1
    student['C']=C_count
  else:
    if 'D' not in student:
      student['D']=0
    D_count+=1
    student['D']=D_count

print(student)
```
