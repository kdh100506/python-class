```python
def lst_create():
  nums=list(map(int,input().split()))
  return nums

def lst_append(lst_):
  while True:
      n=input()
      try:
          lst_.append(float(n))
      except ValueError:
          break
  return lst_
  
def lst_cal(lst_):
  total=0
  for i in lst_:
    total+=i
  avg=total/len(lst_)
  return total, avg

def lst_print(lst_,result_):
  print('점수',lst_)
  print('평균',result_)

#메인
lst=lst_create()
lst=lst_append(lst)
result=lst_cal(lst)
lst_print(lst, result)
```
