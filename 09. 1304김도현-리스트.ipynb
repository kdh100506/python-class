```python
family=['father','mother','brother','sister']
print(family[3])
family.append('puppy')
print(family)
family.append('cat')
print(family)
family.pop() #괄호가 빈칸이면 가장 뒤에것을 삭제
print(family)
family.sort() #괄호가 빈칸이면 오름차순
print(family)
print(family.index('puppy'))
print(len(family))
del family[3]
print(family)
```
