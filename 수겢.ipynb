```python
#1번

point=list(map(int,input().split()))

a=0
for i in point:
  a+=i
b=a/len(point)
c=max(point)
d=min(point)
print(f'총점: {a}')
print(f'평균: {b}')
print(f'최고: {c}')
print(f'최저: {d}')
```

```python
#2번

lst=list(map(int,input().split()))
even=[]
odd=[]

for i in lst:
  if i%2==0:
    even.append(i)
  else:
    odd.append(i)

print(even)
print(odd)
```

```python
#3번

tup=tuple(map(str,input().split()))
tup_lst=[]
for i in tup:
  a=len(i)
  tup_lst.append(a)
b=max(tup_lst)
c=tup_lst.index(b)
print(tup[c])
```

```python
#4번

words=set(map(str,input().split()))
print(len(words))
```

```python
#5번

people=int(input())
dic={}

for i in range(people):
  a,b=map(str,input().split())
  if int(b)>=80:
    b='합격'
  else:
    b='불합격'
  dic[a]=b

for i in dic:
  print(i, dic[i])
```

```python
#6번
set_1=set(map(int,input().split()))
set_2=set(map(int,input().split()))
print(set_1 | set_2)
print(set_1 & set_2)
print(set_1 - set_2)
```

```python
#7번

people=int(input())
dic={}

for i in range(people):
  a,b=map(str,input().split())
  dic[a]=int(b)

point=[]
people_=[]
for i in dic:
  point.append(dic[i])
  people_.append(i)
d=max(point)
e=point.index(d)
print(people_[e],d)


```
