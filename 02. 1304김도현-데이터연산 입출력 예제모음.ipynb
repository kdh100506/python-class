```python
# 8번
a=700
b=500
c=2500

aa=int(input('연필의 판매 개수는?'))
bb=int(input('지우개의 판매 개수는?'))
cc=int(input('공책의 판매 개수는?'))

print('----------------------------')
print('파이썬 문방구의 오늘의 판매 내역')
print('----------------------------')
print('연필:',aa,'개',a*aa,'원')
print('연필:',bb,'개',b*bb,'원')
print('연필:',cc,'개',c*cc,'원')
print('----------------------------')
print('오늘 총 판매 금액:',a*aa+b*bb+c*cc)
print('----------------------------')
```
