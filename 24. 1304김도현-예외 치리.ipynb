```python
a,b=map(int,input().split())
try:
  c=a/b
except:
  print('0으로 나눌 수 없음')
else:
  print(c)
```

```python
a,b=map(int,input().split())
try:
  c=a/b
except ZeroDivisionError as c:
  print('0으로 나눌 수 없음')
else:
  print(c)
```
