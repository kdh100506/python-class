```python
print('점수는 100~0점 까지 입력가능')
point=int(input('점수:'))

if point<=100 and point>=0:
    if point>=90:
        print('A')
    elif point>=80:
        print('B')
    elif point>=70:
        print('C')
    elif point>=60:
        print('D')
    else:
        print('E')
else:
    print('다시 입략하세요.')
    point=int(input('점수:'))
```
